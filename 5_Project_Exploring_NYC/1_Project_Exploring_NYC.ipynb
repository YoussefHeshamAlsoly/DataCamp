# My code

### Which NYC schools have the best math results?


```python
best_math_schools = schools[schools["average_math"]>=640].sort_values("average_math", ascending=False)
best_math_schools = best_math_schools.loc[:, ("school_name", "average_math")].nlargest(10, columns="average_math")
best_math_schools
```

### What are the top 10 performing schools based on the combined SAT scores?

```python
# top_10_schools["total_SAT"] = schools[["average_writing", "average_reading", "average_math"]].sum(axis=1)
top_schools = schools
top_schools["total_SAT"] = schools["average_writing"] + schools["average_reading"] + schools["average_math"]

top_schools = top_schools.loc[:, ("school_name", "total_SAT")]
top_schools = top_schools.sort_values("total_SAT", ascending=False)
top_10_schools = top_schools.nlargest(10, columns="total_SAT")
top_10_schools
```

### Which single borough has the largest standard deviation in the combined SAT score?

```python
SAT = schools
SAT["total_SAT"] = schools[["average_writing", "average_reading", "average_math"]].sum(axis=1)
# SAT

stats_df = SAT.groupby("borough")["total_SAT"].agg(["mean", "std"]).round(2)
stats_df["num_schools"] = SAT.groupby("borough")["school_name"].size()
# stats_df
# stats_df[stats_df['std']==stats_df['std'].max()]

largest_std_dev = stats_df[stats_df['std']==stats_df['std'].max()]
largest_std_dev.rename(columns={"mean":"average_SAT", "std":"std_SAT", "borough":"borough", "num_schools":"num_schools"})

# largest_std_dev = largest_std_dev.set_index("borough", inplace=True)

# largest_std_dev.loc[:, ["borough", "num_schools", "average_SAT", "std_SAT"]]
largest_std_dev
```
