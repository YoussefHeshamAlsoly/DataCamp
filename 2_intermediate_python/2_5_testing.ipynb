```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# numpy and matplotlib imported, seed set.
np.random.seed(123)
```

```python
# initialize and populate all_walks
all_walks = []
for i in range(5) :
    random_walk = [0]
    for x in range(100) :
        step = random_walk[-1]
        dice = np.random.randint(1,7)
        if dice <= 2:
            step = max(0, step - 1)
        elif dice <= 5:
            step = step + 1
        else:
            step = step + np.random.randint(1,7)
        random_walk.append(step)
    all_walks.append(random_walk)
```

```python
# Convert all_walks to NumPy array: np_aw
np_aw = np.array(all_walks)

# Plot np_aw and show
plt.plot(np_aw)
plt.show()

# Clear the figure
plt.clf()
```

```python
print(np_aw)
```

```python
# Transpose np_aw: np_aw_t
np_aw_t = np.transpose(np_aw)

# Plot np_aw_t and show
plt.plot(np_aw_t)
plt.show()
```

```python
# numpy and matplotlib imported, seed set

# Simulate random walk 500 times
all_walks_two = []
for i in range(500) :
    random_walk_two = [0]
    for x in range(100) :
        step_two = random_walk_two[-1]
        dice_two = np.random.randint(1,7)
        if dice_two <= 2:
            step_two = max(0, step_two - 1)
        elif dice_two <= 5:
            step_two = step_two + 1
        else:
            step_two = step_two + np.random.randint(1,7)
        if np.random.rand() <= 0.001 :
            step_two = 0
        random_walk_two.append(step_two)
    all_walks_two.append(random_walk_two)

# Create and plot np_aw_t_two
np_aw_t_two = np.transpose(np.array(all_walks_two))

# Select last row from np_aw_t_two: ends
ends = np_aw_t_two[-1, :]


# Plot histogram of ends, display plot
plt.hist(ends)
plt.show()
```
