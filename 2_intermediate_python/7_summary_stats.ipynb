```python
# Create a DataFrame with 20 rows and some sample data
data = {
    'ID': [i for i in range(1, 21)],
    'Name': [f'Person_{i}' for i in range(1, 21)],
    'Age': np.random.randint(18, 60, size=20),
    'Score': np.random.randint(50, 100, size=20)
}

df = pd.DataFrame(data)
# Display the DataFrame
# df
```

```python
print(df["Age"].median())
print(df["Age"].quantile([.25, .5, .75]))
```

```python
def pct30(column):
    return column.quantile(0.25)

def pct40(column):
    return column.quantile(0.75)
```

```python
# the "agg" (short for aggrigated) method works by applying whatever method that gets passed to it onto
# the column/s that it -meaning the agg method- is called from

print(df["Age"].agg([pct30, pct40]))
print(pct30(df["Age"]), "\t", pct40(df["Age"]))
```
