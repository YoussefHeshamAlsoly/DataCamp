```python
import pandas as pd
import numpy as np
```

```python
# Create a DataFrame with 20 rows and some sample data
data = {
    'ID': [i for i in range(1, 21)],
    'Name': [f'Person_{i}' for i in range(1, 21)],
    'Age': np.random.randint(18, 60, size=20),
    'Score': np.random.randint(50, 100, size=20)
}

df = pd.DataFrame(data)
# Display the DataFrame
# df
```

```python
df.sort_values(["Score", "Age"], ascending=[True, True])
```
